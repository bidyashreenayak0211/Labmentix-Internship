## 3. ***Data Wrangling***

```python
# Handling missing values
df1.drop(columns=['neighbourhood'], inplace=True)
df2.drop(columns=['neighbourhood'], inplace=True)

df1['price'].fillna(df1['price'].median(), inplace=True)
df2['price'].fillna(df2['price'].median(), inplace=True)

df1['last_review'].fillna(pd.NaT, inplace=True)
df2['last_review'].fillna(pd.NaT, inplace=True)

df1['reviews_per_month'].fillna(0, inplace=True)
df2['reviews_per_month'].fillna(0, inplace=True)

df1['host_name'].fillna("Unknown", inplace=True)
df2['host_name'].fillna("Unknown", inplace=True)

df1['license'].fillna("Unknown", inplace=True)
df2['license'].fillna("Unknown", inplace=True)

# Convert data types
df1['last_review'] = pd.to_datetime(df1['last_review'])
df2['last_review'] = pd.to_datetime(df2['last_review'])

df1['price'] = df1['price'].astype(int)
df2['price'] = df2['price'].astype(int)

# Removing outliers
price_cap1 = df1['price'].quantile(0.99)
df1 = df1[df1['price'] <= price_cap1]

price_cap2 = df2['price'].quantile(0.99)
df2 = df2[df2['price'] <= price_cap2]

min_nights_cap1 = df1['minimum_nights'].quantile(0.99)
df1 = df1[df1['minimum_nights'] <= min_nights_cap1]

min_nights_cap2 = df2['minimum_nights'].quantile(0.99)
df2 = df2[df2['minimum_nights'] <= min_nights_cap2]

# Standardizing column names
df1.columns = df1.columns.str.lower().str.replace(" ", "_")
df2.columns = df2.columns.str.lower().str.replace(" ", "_")

# Removing unnamed index column if it exists
df1.drop(columns=[col for col in df1.columns if 'unnamed' in col.lower()], inplace=True)
df2.drop(columns=[col for col in df2.columns if 'unnamed' in col.lower()], inplace=True)
```

```python
# Adding a new column to distinguish the datasets before merging
df1['city'] = 'Chicago'
df2['city'] = 'New Orleans'

# Merging the datasets
df = pd.concat([df1, df2], ignore_index=True)

# Display basic info after merging
print(df.info())

# Checking for missing values in the combined dataset
print(df.isnull().sum())
```

```python
# Merge both cleaned datasets
df = pd.concat([df1, df2], ignore_index=True)

# Save the final merged dataset
df.to_csv("merged_cleaned_airbnb.csv", index=False)
```

#### Chart - 1 Correlation Matrix

##### 1. Why did you pick the specific chart?
I selected a heatmap as it offers a clear visual summary of the relationships between numerical variables. Its use of color gradients makes it easy to spot strong positive or negative correlations, which is useful for detecting multicollinearity and identifying overlapping or unnecessary features. Compared to scatter plots, a heatmap allows for the simultaneous analysis of multiple variable relationships, making it an efficient tool for feature selection and informed data analysis.

##### 2. What is/are the insight(s) found from the chart?
The heatmap highlights significant relationships among numerical variables, making it easier to pinpoint which features are closely linked. #### Chart - 2 Price Distribution

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Price Distribution using KDE Plot
plt.figure(figsize=(8, 5))
sns.kdeplot(data=df, x='price', fill=True, color='skyblue')
plt.xlabel('Price')
plt.ylabel('Density')
plt.title('Price Distribution - KDE Plot')
plt.show()
```

##### 1. Why did you pick the specific chart?

The Kernel Density Estimate (KDE) plot was chosen to visualize the price distribution because it provides a smooth, continuous curve that represents the underlying probability density of the data. Unlike histograms, which can be affected by bin size and create a blocky appearance, KDE plots offer a clearer and more intuitive understanding of how values are distributed across the range. This is especially useful for identifying the shape of the distribution, such as skewness or the presence of multiple peaks, which is common in Airbnb pricing data due to varying property types and locations. The KDE plot helps in analyzing trends more precisely and is visually cleaner, making it ideal for presentations, reports, and identifying patterns in continuous numerical data like price.

##### 2. What is/are the insight(s) found from the chart?

The KDE chart helps in understanding how Airbnb listing prices are distributed by showing whether the data follows a normal pattern or is skewed. It visually indicates if most properties are clustered within a common price range or if there are outliers that deviate significantly from the norm. For instance, a right-skewed distribution would imply that while most listings are priced moderately, there are a few high-end or luxury listings that drive up the overall average. This kind of insight is valuable for identifying pricing trends, spotting anomalies, and understanding how different segments of the market behave. It also helps hosts or analysts recognize how typical or atypical their listing prices are in comparison to the broader dataset.
